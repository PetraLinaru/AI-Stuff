# Understanding Audio Data
## Understanding sound waves


```python
import librosa, librosa.display
import matplotlib.pyplot as plt
import numpy as nps

```

## Waveform Representation

```python
plt.close('all')
file='blues.wav'

##Waveform Representation
signal, sr = librosa.load(file,sr=22050)
#signal = array with values, each equal with sr*T, sr=22050,T=30

#Plot
plt.figure(figsize=(12,4))
librosa.display.waveshow(signal,sr=sr)

```

## Spectrum, Magnitude and Frequency

```python
plt.close('all')
##FFT ->Spectrum
fft = np.fft.fft(signal)

#Magnitude -> Absolute value of all complex values
magnitude=np.abs(fft)

#Frequency
#linespace equally spaced values in range
frequency = np.linspace(0,sr,len(magnitude))

plt.figure()
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.plot(frequency,magnitude)


```

## Interpretation :
Most of the energy is concentrated at the beginning and at the end.
The plot is symmetric and that's a property of the Furrier Series.
We do not need the whole plot therefore, we just need half of the plot, the very first half.

```python
plt.close('all')
left_frequency=frequency[:10000]
left_magnitude=frequency[:10000]
plt.figure()
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.plot(left_frequency,left_magnitude)
plt.show()
```

## Short Time Fourrier Transform to generate a spectogram

```python
number_of_samples=2048 #the window that we are considering when performing a simple Fourrier transform
hop_length=512 #the amounth we are shifting to the right after obtaining an interval from the Fourrier transform
stft = librosa.core.stft(signal, hop_length = hop_length, n_fft = number_of_samples)
spectogram = np.abs(stft)

log_spectogram=librosa.amplitude_to_db(spectogram)

plt.close('all')
#Plotting results
#specshow helps us visualize data as spectogram
plt.figure()
librosa.display.specshow(log_spectogram, sr=sr, hop_length=hop_length)

plt.xlabel('Time')
plt.ylabel('Frequency')

```

## MFCCs


```python
MFCCs=librosa.feature.mfcc(y=signal,n_mfcc=13,hop_length=hop_length)

plt.close('all')
#Plotting results

plt.figure()
librosa.display.specshow(MFCCs, sr=sr, hop_length=hop_length,n_fft=number_of_samples)

plt.xlabel('Time')
plt.ylabel('MFCC')
```
